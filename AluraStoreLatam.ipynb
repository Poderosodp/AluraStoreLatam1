### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Calcular Ingresos por tienda
def calculo_Ingresos(tienda: pd.DataFrame) -> float:
    return tienda['Precio'].sum()

ingresosTienda = calculo_Ingresos(tienda)
ingresosTienda2 = calculo_Ingresos(tienda2)
ingresosTienda3 = calculo_Ingresos(tienda3)
ingresosTienda4 = calculo_Ingresos(tienda4)

#Imprimir resultados
print(f"Ingresos de la Tienda 1: {ingresosTienda}")
print(f"Ingresos de la Tienda 2: {ingresosTienda2}")
print(f"Ingresos de la Tienda 3: {ingresosTienda3}")
print(f"Ingresos de la Tienda 4: {ingresosTienda4}")

listaTiendaIngresos = [ingresosTienda, ingresosTienda2, ingresosTienda3, ingresosTienda4]
listaTiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

colores_hex = ['#A9A9A9', '#8FBC8F', '#778899', '#BC8F8F']
plt.bar(x= listaTiendas, height=listaTiendaIngresos, color=colores_hex)
plt.title('Ingresos por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.show()
```

# 2. Ventas por categoría

```python
# Calcular todas las ventas por categoria
def ventas_X_categoria(tienda_df, nombre_tienda):
    ventas_categoria = tienda_df['Categoría del Producto'].value_counts()

    print(f"\nVentas por categoría en la {nombre_tienda}:\n")
    print(ventas_categoria)

    # Creación del gráfico circular
    plt.figure(figsize=(8, 8))
    plt.pie(ventas_categoria, labels=ventas_categoria.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Ventas por categoría en de la {nombre_tienda}')
    plt.axis('equal')
    plt.show()

# Ejecución de cálculo por tienda
ventas_X_categoria(tienda, 'Tienda 1')
ventas_X_categoria(tienda2, 'Tienda 2')
ventas_X_categoria(tienda3, 'Tienda 3')
ventas_X_categoria(tienda4, 'Tienda 4')
```

# 3. Calificación promedio de la tienda


```python
# Calcular promedio por tienda
def calificacion_promedio(tienda: pd.DataFrame) -> float:
    return round(tienda['Calificación'].mean(), 4)

calificacionTienda = calificacion_promedio(tienda)
calificacionTienda2 = calificacion_promedio(tienda2)
calificacionTienda3 = calificacion_promedio(tienda3)
calificacionTienda4 = calificacion_promedio(tienda4)

print(f'Calificación promedio Tienda 1 = {calificacionTienda}')
print(f'Calificación promedio Tienda 2 = {calificacionTienda2}')
print(f'Calificación promedio Tienda 3 = {calificacionTienda3}')
print(f'Calificación promedio Tienda 4 = {calificacionTienda4}')

listaTiendaCalificacion = [calificacionTienda, calificacionTienda2, calificacionTienda3, calificacionTienda4]
listaTiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
plt.plot(listaTiendas, listaTiendaCalificacion, marker='o')
plt.xlabel('Tienda')
plt.ylabel('Promedio')
plt.title('Calificación promedio por tienda')
plt.grid(True)
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Carcular productos mas y menos vendidos
def product_vendidos_masomenos(tienda_df: pd.DataFrame, nombre_tienda: str):
    ventas_por_producto = tienda_df.groupby('Producto').size().sort_values(ascending=False)

    print(f"\nProductos más vendidos en al {nombre_tienda}:")
    print(ventas_por_producto.head())

    print(f"\nProductos menos vendidos en al {nombre_tienda}:")
    print(ventas_por_producto.tail())

    #Gráfico
    plt.figure(figsize=(12, 6))

    # Productos más vendidos
    ventas_por_producto.head().plot(kind='bar', color='skyblue', ax=plt.gca())
    plt.title(f'Top de los 5 productos más vendidos en la {nombre_tienda}')
    plt.xlabel('Productos')
    plt.ylabel('Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Productos menos vendidos
    plt.figure(figsize=(12, 6))
    ventas_por_producto.tail().plot(kind='bar', color='salmon', ax=plt.gca())
    plt.title(f'Top de los 5 productos menos vendidos en la {nombre_tienda}')
    plt.xlabel('Productos')
    plt.ylabel('Ventas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Cálculo por tienda
product_vendidos_masomenos(tienda, 'Tienda 1')
product_vendidos_masomenos(tienda2, 'Tienda 2')
product_vendidos_masomenos(tienda3, 'Tienda 3')
product_vendidos_masomenos(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
# Calcular costos de envío
def costo_envio(tienda: pd.DataFrame) -> float:
    return round(tienda['Costo de envío'].mean(), 2)

costoEnvioTienda1 = costo_envio(tienda)
costoEnvioTienda2 = costo_envio(tienda2)
costoEnvioTienda3 = costo_envio(tienda3)
costoEnvioTienda4 = costo_envio(tienda4)

print(f'Costo promedio de envío de la Tienda 1 = {costoEnvioTienda1}')
print(f'Costo promedio de envío de la Tienda 2 = {costoEnvioTienda2}')
print(f'Costo promedio de envío de la Tienda 3 = {costoEnvioTienda3}')
print(f'Costo promedio de envío de la Tienda 4 = {costoEnvioTienda4}')

costos_envio = {
    'Tienda 1': costoEnvioTienda1,
    'Tienda 2': costoEnvioTienda2,
    'Tienda 3': costoEnvioTienda3,
    'Tienda 4': costoEnvioTienda4
}

tiendas = list(costos_envio.keys())
promedios = list(costos_envio.values())

plt.figure(figsize=(8, 6))
plt.bar(x = tiendas, height=promedios, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Costo promedio de envío por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envío')
plt.ylim(0, max(promedios) * 1.2)
for i, v in enumerate(promedios):
    plt.text(i, v + 0.1, str(v), ha='center')
plt.show()
```
